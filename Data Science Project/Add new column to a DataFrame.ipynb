## 2. Adding a new column with an array

```python
# import numpy
import numpy as np
```

```python
# create an array of 100 elements
language_score = np.arange(0, 1000)
```

```python
# length of the array
len(np.arange(0, 1000))
```

```python
# adding a new column to the dataframe with an array
df_exams['language score'] = language_score
df_exams
```

```python
# create random integer numbers between 1 and 100
int_language_score = np.random.randint(0, 100, size = 1000)
```

```python
# min value inclusive and high value exclusive
min(int_language_score)
max(int_language_score)
```

```python
# adding a new column to the dataframe with an array
df_exams['language score'] = int_language_score
df_exams
```
