```python
import pandas as pd
```

```python
# reading the csv file
df_pop_raw = pd.read_csv('population_total.csv')
```

```python
# first look at the dataset
df_pop_raw.round(2)
```

# Making a Pivot Table

```python
# .pivot(): Returns a reshaped DataFrame organized by given index / column values ("pivot without aggregation")
```

```python
# dropping null values
df_pop_raw.dropna(inplace = True)
```

```python
# making a pivot table
df_pivot = df_pop_raw.pivot(index = 'year', columns = 'country', values = 'population')
```

```python
# selecting some countries
df_pivot = df_pivot[['United States', 'India', 'China', 'Indonesia', 'Brazil']]
```

```python
# showing pivot table
df_pivot
```

## Lineplot

```python
df_pivot.plot(kind = 'line', xlabel = 'Year', ylabel = 'Billion', title = 'Population Growth (1955-2020)', figsize = (8, 4))
```

## Barplot

```python
# selecting only one one (2020)
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
```

```python
# show new dataframe
df_pivot_2020
```

```python
# transpose dataframe from rows to columns or viceversa
df_pivot_2020 = df_pivot_2020.T
```

```python
# making the barplot
df_pivot_2020.plot(kind = 'bar', color = 'magenta',
                   xlabel = 'Year', ylabel = 'Billion', title = 'Population Growth (2020)')
```

```python
# selecting a few years
df_pivot_sample = df_pivot[df_pivot.index.isin([1980, 1990, 2000, 2010, 2020])]
```

```python
df_pivot_sample
```

```python
# making the grouped barplot
df_pivot_sample.plot(kind = 'bar', xlabel = 'Year', ylabel = 'Billion')
```

## Piechart

```python
# changing column name
df_pivot_2020.rename(columns = {2020:'2020'}, inplace = True)
```

```python
# making the plot
df_pivot_2020.plot(kind = 'pie', y = '2020')
```
