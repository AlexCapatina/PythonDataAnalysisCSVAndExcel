```python
import pandas as pd
```

# pivot()

```python
#read and show dataset
df_gdp = pd.read_csv('gdp.csv', encoding = 'latin-1').round(2)
df_gdp
```

```python
# reshape the dataframe with .pivot()
df_gdp.pivot(index = "year", columns = "country", values = "gdppc")
```

# pivot_table()

```python
# read and show the dataset
df_sales = pd.read_excel("supermarket_sales.xlsx")
df_sales
```

```python
# make a pivot table and add an aggregate function - nu merge
df_sales.pivot_table(values =['Quantity', 'Rating', 'Tax 5%', 'Total', 'Unit price', 'cogs',
                             'gross income', 'gross margin percentage'] , index = "Gender", aggfunc = "sum").round(2)
```

```python
# make a pivot table, add aggregate function, and select some columns
df_sales.pivot_table(values =['Quantity','Total'], index = "Gender", aggfunc = "sum").round(2)
```

```python
# make a pivot table that says how much male and female spent in each section
df_sales.pivot_table(values =['Quantity','Total'],columns = "Product line", index = "Gender", aggfunc = "sum").round(2)
```
